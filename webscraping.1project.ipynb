```python
Model = []

for i in soup.find_all("div", class_="_4rR01T"):
    text = i.text
    regex = re.findall(r"\s(\d+\s\w+)", text)  # Corrected the regular expression pattern
    if regex:
        Model.append(regex[0])
    else:
        Model.append(np.nan)
```

```python
Stars = []

for i in soup.find_all("div", class_="_4rR01T"):
    text = i.text.strip()
    regex = re.search(r"(\d+\s?stars?)", text, re.IGNORECASE)  # Match "X star(s)" with case-insensitivity
    if regex:
        Stars.append(regex.group(1))
    else:
        Stars.append(None)
```

```python
Weights = []

for i in soup.find_all("div", class_="_4rR01T"):
    text = i.text.strip()
    regex = re.search(r"(\d+\.\d+\s?ton)", text, re.IGNORECASE)  # Match "X.X ton" with case-insensitivity
    if regex:
        Weights.append(regex.group(1))
    else:
        Weights.append(None)
```

```python
brand=[]
Model=[]
Stars=[]
Weights=[]
PowerUsage=[]
price=[]
Rating=[]
RoomSize=[]
for i in range(1,21):
    url=f"https://www.flipkart.com/search?q=ac&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    print(url)
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    
    for i in soup.find_all("div",class_="_4rR01T"):
        text=i.text
        regex=re.findall("^\w+",text)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)
            
    for i in soup.find_all("div", class_="_4rR01T"):
        text = i.text
        regex = re.findall(r"\s(\d+\sModel)", text)  # Corrected the regular expression pattern
        if regex:
            Model.append(regex[0])
        else:
            Model.append(np.nan)   
            
    for i in soup.find_all("div", class_="_4rR01T"):
        text = i.text.strip()
        regex = re.search(r"(\d+\s?stars?)", text, re.IGNORECASE)  # Match "X star(s)" with case-insensitivity
        if regex:
            Stars.append(regex.group(1))
        else:
            Stars.append(None)
          
    for i in soup.find_all("div", class_="_4rR01T"):
        text = i.text.strip()
        regex = re.search(r"(\d+\.\d+\s?ton)", text, re.IGNORECASE)  # Match "X.X ton" with case-insensitivity
        if regex:
            Weights.append(regex.group(1))
        else:
            Weights.append(None) 
            
    for i in soup.find_all("div", class_="fMghEO"):
        text = i.text.strip()
        regex = re.search(r"(\d+\.\d+\w+)?\s?kWh", text)  # Match "XW" or "X W"
        if regex:
            PowerUsage.append(regex.group(1))
        else:
            PowerUsage.append(None) 
            
    for i in soup.find_all("div",class_="_2kHMtA"):
        p=i.find("div",class_="_30jeq3 _1_WHN1")
        if p:
            price.append(p.text)
        else:
            price.append(np.nan)  
            
    for i in soup.find_all("div",class_="_2kHMtA"):
        rating=i.find("div",class_="_3LWZlK")
        if rating:
            Rating.append(rating.text)
        else:
            Rating.append(np) 
            
    for i in soup.find_all("div", class_="_2kHMtA"):
        text = i.text.strip()
        room_size_regex = re.search(r"(\d+\s?\d+\s?(sqft))", text, re.IGNORECASE)
        if room_size_regex:
            RoomSize.append(room_size_regex.group(1))
        else:
            RoomSize.append(None)       
             
            
```
